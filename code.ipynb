```python
import numpy as np
from keras.preprocessing import image
import matplotlib.pyplot as plt
from keras.models import load_model

model = load_model('model_fer2013.h5') 

label_dict = {0: 'Angry', 1: 'Disgust', 2: 'Fear', 3: 'Happy', 4: 'Neutral', 5: 'Sad', 6: 'Surprise'}

img_path = "test/fear/PrivateTest_11014592.jpg"  
# img_path = "test/neutral/PrivateTest_60797107.jpg"  
img = image.load_img(img_path, target_size=(48, 48), color_mode="grayscale")
img_array = np.array(img)

img = np.expand_dims(img_array, axis=0)  
img = img.reshape(1, 48, 48, 1)

result = model.predict(img)
result = list(result[0])
predicted_label = label_dict[result.index(max(result))]

plt.imshow(img_array, cmap='gray')  
plt.axis('off')  
plt.title(f"Predicted Emotion: {predicted_label}")  
plt.show()

print(f"Predicted Emotion: {predicted_label}")
```
